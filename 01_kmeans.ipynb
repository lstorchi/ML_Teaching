```python
import matplotlib.pyplot as plt
import random

x = []
y = []

random.seed(0)

for i in range(10):
    x.append(random.randint(1, 30))
    y.append(random.randint(1, 30))

print(x)
print(y)

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

data = list(zip(x, y))
inertias = []
silhouettes = []
for i in range(2,9):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(data, 
                    kmeans.labels_))

plt.plot(range(2,9), inertias, marker='o')
plt.title('KMeans Inertia')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()  
plt.plot(range(2,9), silhouettes, marker='o')
plt.title('KMeans Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

print(kmeans.labels_)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

```python
# plots also centroids of clusters
print(kmeans.cluster_centers_)

plt.scatter(x, y, c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], \
            kmeans.cluster_centers_[:, 1], \
            c='red')
plt.show()
```
