```python
import re
import numpy as np
import pandas as pd

# 1 - Optimize the hyperparameters of the models, both RF and NN
# 2 - Use all the conformers, see code below

df = pd.read_csv('BBB_features.csv', sep=';')
label = pd.read_csv('BBB_labels.txt', header=None, sep=' ')
logbb = []
permeate = []
for molname in df['Objects']:
    basename = re.sub(r'_c\d+', '', molname)
    if basename in label[0].values:
        lobBB = label[label[0] == basename][1].values[0]
        logbb.append(lobBB)
        if lobBB > 0:
            permeate.append(1)
        else:
            permeate.append(0)
    else:
        print('Warning: cannot find ', molname)
print('LogBB:', len(logbb))
print('Permeate:', len(permeate))
print('Features:', df.shape)
```
